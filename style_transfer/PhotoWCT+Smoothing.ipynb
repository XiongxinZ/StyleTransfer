```python
import matplotlib.pyplot as plt
%matplotlib inline

mark = ['-o','-.','-v','-s','-x', '-8']
process = ['stylization', 'propagation', 'post processing']
xlabel = ['rococo', 'abstract', 'cubism', 'impressionism', 'baroque', 'abstract1']

ets_1 = [[0.294302, 0.307188, 0.298324, 0.317484, 0.286072, 0.268372],
       [4.548169, 4.586918, 4.604106, 4.547896, 4.599716, 4.578470],
       [0.194853, 0.194189, 0.199021, 0.193089, 0.200598, 0.196182]]


def draw(ets):
  for i in range(len(ets)):
    plt.plot(xlabel, ets[i], mark[i])
    # plt.rcParams["figure.figsize"] = (40,30)
    plt.legend(process[i:i+1])
    plt.xlabel('Style', fontsize=18)
    plt.ylabel('Elapsed time', fontsize=16)

    plt.show()

def draw1(ets):
  for i in range(len(ets)):
    plt.plot(xlabel, ets[i], mark[i])
    # plt.rcParams["figure.figsize"] = (40,30)
    plt.legend(process)
    plt.xlabel('Style', fontsize=18)
    plt.ylabel('Elapsed time', fontsize=16)

  plt.show()

draw(ets_1)
draw1(ets_1)
```
