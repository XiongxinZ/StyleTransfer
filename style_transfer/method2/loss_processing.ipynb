```python
def get_loss(file_name):
    loss_file = open(file_name)
    output = []
    for line in loss_file:
        output.append(line.strip())
    output = output[4:-1:5]
    loss = []
    #improvement = []
    for item in output:
        s = item.split(' ')
        loss.append(s[3])
        #improvement.append(s[6])
    loss = [float(l) for l in loss]
    #improvement = [float(i) for i in improvement]
    #return loss, improvement
    return loss
```

```python
ab2_loss = get_loss('loss_output/front_ab2.txt')
abs_loss = get_loss('loss_output/front_abstract.txt')
baroque_loss = get_loss('loss_output/front_baroque.txt')
cubism_loss = get_loss('loss_output/front_cubism.txt')
impressionism_loss = get_loss('loss_output/front_impression.txt')
rococo_loss = get_loss('loss_output/front_rococo.txt')
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
x = [i for i in range(1,31)]
plt.plot(x, ab2_loss)
plt.plot(x, abs_loss)
plt.plot(x, baroque_loss)
plt.plot(x, cubism_loss)
plt.plot(x, impressionism_loss)
plt.plot(x, rococo_loss)
plt.legend(["abstract1", "abstract2", "baroque", "cubism", "impressionism", "rococo"])
plt.xlabel('iterations')
plt.ylabel('loss')
```
